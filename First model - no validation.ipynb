```python
df = pd.read_csv('driver_imgs_list.csv')
df.head()
```

```python
img_list = pd.read_csv(path/'driver_imgs_list.csv')

# select a subset of the subjects for validation
valid_subjects = img_list.subject.sort_values().unique()[-4:]
# create new column identifying the subjects for validation
img_list['is_valid'] = img_list['subject'].isin(valid_subjects)

print("valid subjects: ", valid_subjects)
print(img_list[img_list['is_valid']==True].subject.count())

# create new column with class folder and image name combined
img_list['img_path'] = img_list.classname + '/' + img_list.img

valid_names = img_list[img_list['subject'].isin(valid_subjects)].img
valid_names = valid_names.to_list()
```

```python
batch_tfms = aug_transforms(size=244, do_flip=False)
item_tfms = Resize(480)
```

```python
dls = ImageDataLoaders.from_folder(path/'imgs/train', valid_pct=0.2, batch_tfms=batch_tfms)
dls.show_batch()
```

```python
learn = cnn_learner(dls, resnet34, metrics=accuracy)
```

```python
learn.fit_one_cycle(6, max_lr=1e-2)
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.save('beforeUnFreeze')
```

```python
learn.fit_one_cycle(6, slice(3e-7, 3e-5))
```

```python
learn.save('before')
```

```python
learn.load('beforeUnFreeze')
```

```python
learn.unfreeze()
learn.lr_find()
```

```python
learn.fit_one_cycle(6, slice(5e-7, 5e-4))
```

```python
learn.save('final')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_top_losses(6)
```

```python
images = get_files(path/'imgs/test')
dataBlock = DataBlock()
test = dataBlock.datasets(images)
```

```python
ImageDataLoaders.from_lists(test)
```

```python
test_dl = learn.dls.test_dl(images)
```

```python
preds = learn.get_preds(dl=test_dl)
```

```python
str(test_dl.dataset.items[0])[10:]
```

```python
labels = pd.DataFrame(test_dl.dataset.items, columns=['img'])
labels.img = labels.img.astype(str)
labels = labels.img.str.rsplit('/', 1, expand=True)
labels.drop(0, axis=1, inplace=True)
labels.rename(columns={1: 'img'}, inplace=True)
```

```python
labels.head()
```

```python
columns = dls.vocab

submission = pd.DataFrame(preds[0].numpy(), columns=columns, index=labels.img)
submission.reset_index(inplace=True)
submission.to_csv('submission.csv', index=False)
```
