# Getting the data

```python
path = Path('.')
path
```

```python
Path.BASE_PATH = path
```

```python
path.ls()
```

```python
df = pd.read_csv(path/'driver_imgs_list.csv')
df.head()
```

```python
def add_valid(df, n_valid=4):
  # validation subjects
  valid_subjects = df.sample(frac=1).subject.unique()[-n_valid:]
  # Identifying the validation subjects
  df['is_valid'] = df['subject'].isin(valid_subjects)

  print('Valid Subjects', valid_subjects)
  print((df['is_valid'] == True).sum())

  # adding the path to the df
  df['img_path'] = df['classname'] + '/' + df['img']
  return df
```

```python
n_df = add_valid(df, n_valid=5)
n_df.sample(5)
```

```python
# examine images
img = PILImage.create(path/'imgs/train'/n_df.img_path[111])
print(img.shape)
img.reshape(200, 200)
```

```python
# transforms
item_tfms = Resize(224)
batch_tfms = aug_transforms(size=224, do_flip=False)
```

```python
db = DataBlock(blocks = (ImageBlock, CategoryBlock),
               get_x=ColReader('img_path', pref=path/'imgs/train/'),
               splitter=ColSplitter(),
               get_y=ColReader('classname'),
               batch_tfms=Resize(128))
```

```python
dls = db.dataloaders(df, bs=64)
```

```python
dls.show_batch()
```

```python
len(dls.train_ds), len(dls.valid_ds)
```

```python
learn = cnn_learner(dls, resnet34, metrics=accuracy)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(epochs=6, lr_base=1e-3, freeze_epochs=4)
```

```python
learn.save('128-noaug')
```

```python
# transforms
batch_tfms = aug_transforms(size=128, do_flip=False)
```

```python
db = DataBlock(blocks = (ImageBlock, CategoryBlock),
               get_x=ColReader('img_path', pref=path/'imgs/train/'),
               splitter=ColSplitter(),
               get_y=ColReader('classname'),
               batch_tfms=batch_tfms)
dls = db.dataloaders(df, bs=64)
```

```python
learn = cnn_learner(dls, resnet34, metrics=accuracy)
learn.fine_tune(epochs=6, lr_base=1e-3, freeze_epochs=4)
```

```python
learn.save('128-aug')
```

```python
# training128-noaug + bigger size
db = DataBlock(blocks = (ImageBlock, CategoryBlock),
               get_x=ColReader('img_path', pref=path/'imgs/train/'),
               splitter=ColSplitter(),
               get_y=ColReader('classname'),
               batch_tfms=Resize(224))

dls = db.dataloaders(df, bs=64)
```

```python
# loading the model
learn = cnn_learner(dls, resnet34, metrics=accuracy)
learn.load('128-noaug')
```

```python
learn.dls = dls
```

```python
learn.fine_tune(epochs=8, base_lr=1e-3, freeze_epochs=4)
```

```python
learn.save('128-224')
```

# Testing 

```python
test_images = get_files(path/'imgs/test')
test_dl = learn.dls.test_dl(test_images)
```

```python
preds = learn.get_preds(dl=test_dl)
```

```python
labels = pd.DataFrame(test_dl.dataset.items, columns=['img'])
labels.img = labels.img.astype(str)
labels = labels.img.str.rsplit('/', 1, expand=True)
labels.drop(0, axis=1, inplace=True)
labels.rename(columns={1: 'img'}, inplace=True)
```

```python
columns = dls.vocab

submission = pd.DataFrame(preds[0].numpy(), columns=columns, index=labels.img)
submission.reset_index(inplace=True)
submission.to_csv('submission.csv', index=False)
```
